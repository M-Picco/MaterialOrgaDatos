```python
def correct_duration(x):
    return float(str(x).replace(' ', '').replace('`', '').strip())
df['duration (seconds)'] = df['duration (seconds)'].apply(correct_duration)
```

```python
def correct_datetime(x):
    fecha, hora = x.split(' ')
    hora, minuto = hora.split(':')
    if hora == '24':
        hora = '00'
    return '{0} {1}:{2}'.format(fecha, hora, minuto)

df['datetime'] = df['datetime'].apply(correct_datetime)
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
# El avistaje mas largo ocurrido en enero del 2005 ocurrió en la ciudad de Green Lake
df.loc[(df['datetime'].dt.month == 1) & (df['datetime'].dt.year == 2005)].sort_values(by="duration (seconds)", ascending=False).head(5)
```
